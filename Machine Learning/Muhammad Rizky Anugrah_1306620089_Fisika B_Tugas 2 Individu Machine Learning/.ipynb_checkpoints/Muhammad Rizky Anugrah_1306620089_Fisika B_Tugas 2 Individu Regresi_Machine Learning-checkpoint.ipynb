```python
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=42)
```

```python
class LinearRegressionGD:
    def __init__(self, eta=0.001, n_iter=1000):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        self.w_ = np.zeros(2)
        self.errors_ = []

        for i in range(self.n_iter):
            errors = []
            for xi, yi in zip(X, y):
                error = yi - self.predict(xi)
                errors.append(error)
                self.w_[0] += self.eta * error
                self.w_[1:] += self.eta * error * xi
            mse = np.mean(np.square(errors))
            self.errors_.append(mse)
        return self

    def predict(self, X):
        return self.w_[0] + self.w_[1] * X
    
    def get_params(self):
        return self.w_[0], self.w_[1]
```

```python
lr = LinearRegressionGD()
lr.fit(X_train, y_train)
w0, w1 = lr.get_params()
print("Nilai Intercept: w0 = {}".format(w0))
print("Nilai Koefisien Regresi: w1 = {}".format(w1))
```

```python
print(f"f(x) = {w0:} + {w1:}x")
```

```python
# Calculate bias, variance, and MSE on test dataset
y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

mse = np.mean((y_pred_test - y_test) ** 2)
bias = np.mean((y_pred_test - np.mean(y_test)) ** 2)
variance = np.var(y_pred_test)

# Print bias, variance, and MSE
print('Bias: ', bias)
print('Variance: ', variance)
print('MSE: ', mse)
```
